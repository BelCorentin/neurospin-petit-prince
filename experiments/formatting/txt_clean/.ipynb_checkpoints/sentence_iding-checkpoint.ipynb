```python
import pandas as pd
import numpy as np
from text_to_num import text2num
```

```python
df = pd.read_csv('./MAU_events_chapter_1.csv')
```

```python
df
```

```python
chr_df = pd.read_csv('allwords.csv',sep='\t',names = ['word', 'sentence_id'])
chr_df
```

```python
count_chr = 0
count_df = df.query("kind=='word'")["onset"].idxmin()
for index, row in df.query("kind=='word'").iterrows():

    word =  df.at[count_df, 'word']
    chr_word = chr_df.at[count_chr,'word']

    if word == chr_word:
        df.at[count_df, 'sentence_id'] = chr_df.at[count_chr,'sentence_id']  # TOFIX sentence_id?
        count_df +=1
        count_chr += 1
    
    elif word in chr_word:  #  "l", "avais" in "l'avais"
        df.at[count_df, 'sentence_id'] = chr_df.at[count_chr,'sentence_id']  # TOFIX sentence_id?
        
        count_df +=1
        word =  df.at[count_df, 'word']
        assert word in chr_word
        df.at[count_df, 'sentence_id'] = chr_df.at[count_chr,'sentence_id']  # TOFIX sentence_id?
        
        count_df += 1
        count_chr += 1
    elif text2num(word, lang='fr') == chr_word:
        df.at[count_df, 'sentence_id'] = chr_df.at[count_chr,'sentence_id']  # TOFIX sentence_id?
        count_df +=1
        count_chr += 1
        
    else:
        
        print(word)
        print(chr_word)
        raise ValueError
        

        
```
